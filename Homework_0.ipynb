* View image
* Crop image
* Edit brightness of image
* Rotate image
* Save modified image

```python
import cv2

import os

import numpy as np

import matplotlib.pyplot as plt

#loading in the greyscale image of Apollo 11 with funciton imread
greyscale_Apollo_11 = cv2.imread(os.path.relpath('Data/Apollo_11_Launch.jpg'),cv2.IMREAD_GRAYSCALE)

#viewing image using function imshow and color map is set to grey scale
plt.imshow(greyscale_Apollo_11, cmap="grey")
plt.title("Original Greyscale Image")

#cropping to show the object closely
new_greyscale_Apollo_11 = greyscale_Apollo_11[50:650,400:850]

#create a constant matrix to subtract from the image to modify the brightness of the image
matrix = np.ones(new_greyscale_Apollo_11.shape, dtype="uint8")*50

#subtract the matrix from the newly cropped image
darker_greyscale_Apollo_11 = cv2.subtract(new_greyscale_Apollo_11, matrix)

#rotate the darkened image by 90 degrees with the rotate function 
rotated_greyscale_Apollo_11 = cv2.rotate(darker_greyscale_Apollo_11,cv2.ROTATE_90_CLOCKWISE)

#save the new image as 'Modified_Apollo11_Image using the function imwrite
cv2.imwrite('Modified_Apollo11_Image.jpg',rotated_greyscale_Apollo_11)

```
